```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import csv
import random
from citipy import citipy
import scipy.stats as st
```

```python
from config import api_key

lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:

        cities.append(city)

len(cities)
```

```python
town= "london"
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}&units={units}&q={town}&appid={api_key}"
print(query_url)
```

```python
plt.scatter(weather_data["Lat"], weather_data["Max Temp"], marker="x", s=5)

latt = weather_data["Lat"]
maxt = weather_data["Max Temp"]

slope, int, r, p, std_err = st.linregress(latt, maxt)
fit = slope * latt + int
plt.plot(latt,fit,"-")


# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Max. Temperature")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Weather_CSV/Max_Temp_vs_Latitude.png")
```
